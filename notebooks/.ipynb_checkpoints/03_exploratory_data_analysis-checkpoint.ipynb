```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/student_habits_cleaned.csv")

df.head()
```

```python
df.shape
```

```python
##Target variable distribution
plt.figure(figsize=(8, 5))
plt.hist(df["exam_score"], bins=20)
plt.title("Distribution of Exam Scores")
plt.xlabel("Exam Score")
plt.ylabel("Number of Students")
plt.show()
```

```python
##Correlation with exam score

numeric_cols = df.select_dtypes(include=["int64", "float64"])

correlation = numeric_cols.corr()["exam_score"].sort_values(ascending=False)

correlation
```

```python
correlation.drop("exam_score").plot(kind="bar", figsize=(10, 5))

plt.title("Correlation of Features with Exam Score")
plt.xlabel("Feature")
plt.ylabel("Correlation")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["study_hours_per_day"], df["exam_score"])
plt.title("Study Hours vs Exam Score")
plt.xlabel("Study Hours per Day")
plt.ylabel("Exam Score")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["attendance_percentage"], df["exam_score"])
plt.title("Attendance Percentage vs Exam Score")
plt.xlabel("Attendance Percentage")
plt.ylabel("Exam Score")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["social_media_hours"], df["exam_score"])
plt.title("Social Media Hours vs Exam Score")
plt.xlabel("Social Media Hours")
plt.ylabel("Exam Score")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["sleep_hours"], df["exam_score"])
plt.title("Sleep Hours vs Exam Score")
plt.xlabel("Sleep Hours")
plt.ylabel("Exam Score")
plt.show()
```

```python
df.groupby("diet_quality")["exam_score"].mean().sort_values(ascending=False)
```

```python
diet_avg = df.groupby("diet_quality")["exam_score"].mean().sort_values(ascending=False)

diet_avg.plot(kind="bar", figsize=(7, 5))

plt.title("Average Exam Score by Diet Quality")
plt.xlabel("Diet Quality")
plt.ylabel("Average Exam Score")
plt.xticks(rotation=0)
plt.show()
```

```python
internet_avg = df.groupby("internet_quality")["exam_score"].mean().sort_values(ascending=False)

internet_avg.plot(kind="bar", figsize=(7, 5))

plt.title("Average Exam Score by Internet Quality")
plt.xlabel("Internet Quality")
plt.ylabel("Average Exam Score")
plt.xticks(rotation=0)
plt.show()
```

```python
job_avg = df.groupby("part_time_job")["exam_score"].mean()

job_avg.plot(kind="bar", figsize=(7, 5))

plt.title("Average Exam Score by Part-Time Job Status")
plt.xlabel("Part-Time Job")
plt.ylabel("Average Exam Score")
plt.xticks(rotation=0)
plt.show()
```

```python
## EDA Findings

### Positive Factors
- Study hours per day showed the strongest positive relationship with exam scores (0.825).
- Mental health rating was moderately associated with better academic performance.
- Exercise frequency and sleep hours showed small positive relationships.

### Negative Factors
- Social media usage and Netflix usage showed negative correlations with exam scores.
- Students spending more time on entertainment activities generally achieved lower scores.

### Additional Insights
- Internet quality and part-time employment showed minimal influence on academic performance.
- Diet quality showed only minor differences in average exam scores.

### Conclusion
Study habits and mental well-being appear to be the strongest indicators of student performance in this dataset.
```
