```python
import pandas as pd

df = pd.read_csv("../data/raw/student_habits_performance.csv")

df.head()
```

```python
df.isnull().sum()
```

```python
df["parental_education_level"].value_counts()
```

```python
mode_value = df["parental_education_level"].mode()[0]

df["parental_education_level"] = df["parental_education_level"].fillna(mode_value)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df["student_id"].nunique()
```

```python
df = df.drop("student_id", axis=1)
```

```python
df.shape
```

```python
df.to_csv(
    "../data/processed/student_habits_cleaned.csv",
    index=False
)
```

```python
## Data Cleaning Summary

- Filled 91 missing values in parental_education_level using mode imputation.
- Checked and removed duplicates if present.
- Removed student_id because it has no predictive value.
- Saved cleaned dataset for EDA and machine learning.
```
