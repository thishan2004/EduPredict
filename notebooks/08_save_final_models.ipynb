```python

import pandas as pd
import joblib

# Load processed features and targets
X = pd.read_csv("../data/processed/features_encoded.csv")
y_regression = pd.read_csv("../data/processed/target_regression.csv").squeeze()
y_classification = pd.read_csv("../data/processed/target_classification.csv").squeeze()

print("Features:", X.shape)
print("Regression Target:", y_regression.shape)
print("Classification Target:", y_classification.shape)
```

```python
# Import train_test_split
from sklearn.model_selection import train_test_split

# Regression split
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X,
    y_regression,
    test_size=0.2,
    random_state=42
)

# Classification split
X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(
    X,
    y_classification,
    test_size=0.2,
    random_state=42,
    stratify=y_classification
)
```

```python
# Import final selected models
from sklearn.linear_model import LinearRegression, LogisticRegression

# Final regression model
# Best model for predicting exact exam score
final_regression_model = LinearRegression()

# Final classification model
# Best model for predicting performance category
final_classification_model = LogisticRegression(max_iter=1000)

# Train models
final_regression_model.fit(X_train_reg, y_train_reg)
final_classification_model.fit(X_train_cls, y_train_cls)

print("Final models trained successfully!")
```

```python
# Import metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np

# Regression prediction
y_pred_reg = final_regression_model.predict(X_test_reg)

# Classification prediction
y_pred_cls = final_classification_model.predict(X_test_cls)

# Regression metrics
mae = mean_absolute_error(y_test_reg, y_pred_reg)
rmse = np.sqrt(mean_squared_error(y_test_reg, y_pred_reg))
r2 = r2_score(y_test_reg, y_pred_reg)

# Classification metrics
accuracy = accuracy_score(y_test_cls, y_pred_cls)
precision = precision_score(y_test_cls, y_pred_cls, average="weighted")
recall = recall_score(y_test_cls, y_pred_cls, average="weighted")
f1 = f1_score(y_test_cls, y_pred_cls, average="weighted")

print("Final Regression Model Performance")
print("MAE:", mae)
print("RMSE:", rmse)
print("R2 Score:", r2)

print("\nFinal Classification Model Performance")
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

```python
# Save final regression model
joblib.dump(
    final_regression_model,
    "../models/final_exam_score_model.pkl"
)

# Save final classification model
joblib.dump(
    final_classification_model,
    "../models/final_performance_category_model.pkl"
)

print("Final models saved successfully!")
```

```python
# Save feature column names
feature_columns = list(X.columns)

joblib.dump(
    feature_columns,
    "../models/feature_columns.pkl"
)

print("Feature columns saved successfully!")
```

```python
# Load saved models
loaded_regression_model = joblib.load("../models/final_exam_score_model.pkl")
loaded_classification_model = joblib.load("../models/final_performance_category_model.pkl")
loaded_feature_columns = joblib.load("../models/feature_columns.pkl")

print("Saved models loaded successfully!")
print("Number of feature columns:", len(loaded_feature_columns))
```

```python
# Select one sample student
sample_student = X.iloc[[0]]

# Predict exam score
predicted_score = loaded_regression_model.predict(sample_student)[0]

# Predict performance category
predicted_category = loaded_classification_model.predict(sample_student)[0]

print("Predicted Exam Score:", round(predicted_score, 2))
print("Predicted Performance Category:", predicted_category)
```
