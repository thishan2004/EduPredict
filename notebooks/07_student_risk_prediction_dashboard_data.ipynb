```python

import pandas as pd
import numpy as np

# Load feature-engineered dataset
df = pd.read_csv("../data/processed/student_habits_feature_engineered.csv")

# Display first few rows
df.head()
```

```python
# Check dataset shape and columns
print("Dataset Shape:", df.shape)

# Display all columns
df.columns
```

```python
# Create academic risk level based on exam score
# This will be useful for dashboard reporting

def assign_risk_level(score):
    if score >= 80:
        return "Low Risk"
    elif score >= 60:
        return "Medium Risk"
    else:
        return "High Risk"

# Apply function to create new column
df["academic_risk_level"] = df["exam_score"].apply(assign_risk_level)

# Check distribution of risk levels
df["academic_risk_level"].value_counts()
```

```python
# Create personalized recommendations based on student habits
# This function checks each student's lifestyle and academic behavior

def generate_recommendations(row):
    
    recommendations = []
    
    # Study hours recommendation
    if row["study_hours_per_day"] < 2:
        recommendations.append("Increase daily study hours to at least 2-3 hours.")
    
    # Social media recommendation
    if row["social_media_hours"] > 3:
        recommendations.append("Reduce social media usage to improve focus.")
    
    # Netflix / entertainment recommendation
    if row["netflix_hours"] > 2:
        recommendations.append("Reduce entertainment screen time.")
    
    # Attendance recommendation
    if row["attendance_percentage"] < 75:
        recommendations.append("Improve class attendance to at least 80%.")
    
    # Sleep recommendation
    if row["sleep_hours"] < 6:
        recommendations.append("Improve sleep routine and aim for 7-8 hours of sleep.")
    
    # Mental health recommendation
    if row["mental_health_rating"] < 5:
        recommendations.append("Focus on mental well-being and seek support if needed.")
    
    # Exercise recommendation
    if row["exercise_frequency"] < 2:
        recommendations.append("Increase weekly physical activity for better balance.")
    
    # Diet recommendation
    if row["diet_quality"] == "Poor":
        recommendations.append("Improve diet quality to support better academic performance.")
    
    # Internet quality recommendation
    if row["internet_quality"] == "Poor":
        recommendations.append("Improve internet access or use campus resources for online learning.")
    
    # If no issues are detected
    if len(recommendations) == 0:
        recommendations.append("Maintain current positive study and lifestyle habits.")
    
    return recommendations
```

```python
# Generate recommendations for every student
df["recommendations"] = df.apply(generate_recommendations, axis=1)

# Display sample results
df[["exam_score", "academic_risk_level", "recommendations"]].head()
```

```python
# Create a simplified student report for dashboard use

student_report = df[
    [
        "age",
        "gender",
        "study_hours_per_day",
        "social_media_hours",
        "netflix_hours",
        "attendance_percentage",
        "sleep_hours",
        "mental_health_rating",
        "exercise_frequency",
        "diet_quality",
        "internet_quality",
        "exam_score",
        "academic_risk_level",
        "recommendations"
    ]
]

# View report
student_report.head()
```

```python
# Count students in each academic risk level

risk_summary = df["academic_risk_level"].value_counts().reset_index()

risk_summary.columns = ["Risk Level", "Student Count"]

risk_summary
```

```python
# Analyze average habits for each risk group

risk_habit_summary = df.groupby("academic_risk_level")[
    [
        "study_hours_per_day",
        "social_media_hours",
        "netflix_hours",
        "attendance_percentage",
        "sleep_hours",
        "mental_health_rating",
        "exercise_frequency",
        "exam_score"
    ]
].mean().round(2)

risk_habit_summary
```

```python
# Save reports for the Streamlit dashboard

student_report.to_csv(
    "../data/processed/student_risk_report.csv",
    index=False
)

risk_summary.to_csv(
    "../data/processed/risk_summary.csv",
    index=False
)

risk_habit_summary.to_csv(
    "../data/processed/risk_habit_summary.csv"
)

print("Student risk reports saved successfully!")
```

```python
# This function will be used later in the Streamlit dashboard
# It generates recommendations for a new student profile

def predict_student_recommendations(
    study_hours_per_day,
    social_media_hours,
    netflix_hours,
    attendance_percentage,
    sleep_hours,
    mental_health_rating,
    exercise_frequency,
    diet_quality,
    internet_quality
):
    
    recommendations = []
    
    if study_hours_per_day < 2:
        recommendations.append("Increase daily study hours to at least 2-3 hours.")
    
    if social_media_hours > 3:
        recommendations.append("Reduce social media usage to improve focus.")
    
    if netflix_hours > 2:
        recommendations.append("Reduce entertainment screen time.")
    
    if attendance_percentage < 75:
        recommendations.append("Improve class attendance to at least 80%.")
    
    if sleep_hours < 6:
        recommendations.append("Improve sleep routine and aim for 7-8 hours of sleep.")
    
    if mental_health_rating < 5:
        recommendations.append("Focus on mental well-being and seek support if needed.")
    
    if exercise_frequency < 2:
        recommendations.append("Increase weekly physical activity for better balance.")
    
    if diet_quality == "Poor":
        recommendations.append("Improve diet quality to support better academic performance.")
    
    if internet_quality == "Poor":
        recommendations.append("Improve internet access or use campus resources for online learning.")
    
    if len(recommendations) == 0:
        recommendations.append("Maintain current positive study and lifestyle habits.")
    
    return recommendations
```

```python
# Test the function with a sample student profile

test_recommendations = predict_student_recommendations(
    study_hours_per_day=1,
    social_media_hours=4,
    netflix_hours=3,
    attendance_percentage=65,
    sleep_hours=5,
    mental_health_rating=4,
    exercise_frequency=1,
    diet_quality="Poor",
    internet_quality="Average"
)

print("Recommendations for sample student:")

for rec in test_recommendations:
    print("-", rec)
```
